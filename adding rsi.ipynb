```python
import pandas as pd
from datetime import date
from nsepy import get_history
import os

from pathlib import Path
proj_dir = str(Path.cwd())
```

```python
all_stocks_path = proj_dir + "\\data\\historical\\"
```

```python
def compute_rsi(data, time_window):
    diff = data.diff(1).dropna()
    up_chg = 0 * diff
    down_chg = 0 * diff
    
    up_chg[diff > 0] = diff[ diff>0 ]
    down_chg[diff < 0] = diff[ diff < 0 ]
    
    up_chg_avg   = up_chg.ewm(com=time_window-1 , min_periods=time_window).mean()
    down_chg_avg = down_chg.ewm(com=time_window-1 , min_periods=time_window).mean()
    
    rs = abs(up_chg_avg/down_chg_avg)
    rsi = 100 - 100/(1+rs)
    return rsi
```

```python
for stock in os.listdir(all_stocks_path):
    stock_file = all_stocks_path + stock
    print(stock_file)
    df = pd.read_csv(stock_file)

    try:
        df.drop(["Symbol", "Series", "Turnover", "Trades", "Deliverable Volume", "%Deliverble"], axis = 1, inplace = True)
    except:
        pass
    rsi = compute_rsi(df['Close'], 14)

    df['rsi'] = rsi
    print("RSI added in " + stock_file)
    df.dropna(axis = 0, inplace=True)
    try:
        df.drop(['Unnamed: 0'], axis = 1, inplace = True)
    except:
        pass
    try:
        df.drop(['Unnamed: 0.1'], axis = 1, inplace = True)
    except:
        pass
    try:
        df.drop(['Unnamed: 0.1.1'], axis = 1, inplace = True)
    except:
        pass
    df = df.reset_index(drop=True)
    df.head()
    df.shape
    df.to_csv(stock_file)
```
