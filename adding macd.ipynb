```python
import pandas as pd
from datetime import date
from nsepy import get_history
import os

from pathlib import Path
proj_dir = str(Path.cwd())
```

```python
all_stocks_path = proj_dir + "\\data\\historical\\"
```

```python
def compute_macd(data, span1, span2):
    exp1 = data.ewm(span=span1).mean()
    exp2 = data.ewm(span=span2).mean()
    macd = exp1 - exp2
    return macd
```

```python
def compute_signal(macd, span):
    signal = macd.ewm(span=span).mean()
    return signal
```

```python
for stock in os.listdir(all_stocks_path):
    stock_file = all_stocks_path + stock
    print(stock_file)
    df = pd.read_csv(stock_file)

    try:
        df.drop(["Symbol", "Series", "Turnover", "Trades", "Deliverable Volume", "%Deliverble"], axis = 1, inplace = True)
    except:
        pass

    macd = compute_macd(df['Close'], 12, 26)
    signal = compute_signal(macd, 9)
    df['macd'] = macd
    df['signal'] = signal

    print("MACD added in " + stock_file)
    df.dropna(axis = 0, inplace=True)
    try:
        df.drop(['Unnamed: 0'], axis = 1, inplace = True)
    except:
        pass
    df = df.reset_index(drop=True)
    df.to_csv(stock_file)
```

```python
import matplotlib.pyplot as plt

macd.plot(label='TCS MACD', color='g')
ax = signal.plot(label='Signal Line', color='r')
df['Close'].tail(100).plot(ax=ax, secondary_y=True, label='TCS')
```
