```python
import numpy as np
import pandas as pd
from pathlib import Path
import os

proj_dir = str(Path.cwd())
proj_dir
```

```python
all_stocks_path = proj_dir + "\\data\\historical\\"
```

```python
def add_prev_close_prices(df):
    df['1_lag'] = df.shift(1)['Close']
    df['2_lag'] = df.shift(2)['Close']
    df['3_lag'] = df.shift(3)['Close']
    df['4_lag'] = df.shift(4)['Close']
    df['5_lag'] = df.shift(5)['Close']
    df['6_lag'] = df.shift(6)['Close']
    df['7_lag'] = df.shift(7)['Close']
    df['8_lag'] = df.shift(8)['Close']
    df['9_lag'] = df.shift(9)['Close']
    df['10_lag'] = df.shift(10)['Close']
    df['11_lag'] = df.shift(11)['Close']
    df['12_lag'] = df.shift(12)['Close']
    df['13_lag'] = df.shift(13)['Close']
    df['14_lag'] = df.shift(4)['Close']
    df['15_lag'] = df.shift(15)['Close']
    df['16_lag'] = df.shift(16)['Close']
    df['17_lag'] = df.shift(17)['Close']
    df['18_lag'] = df.shift(18)['Close']
    df['19_lag'] = df.shift(19)['Close']
    df['20_lag'] = df.shift(20)['Close']
    df.dropna(how='any', axis = 0, inplace=True)
    df.reset_index(drop=True)
    return df
```

```python
def lengthOfLIS(nums):
        dp = []
        for i in range(len(nums)):
            dp.append(0)
            for j in range(i):
                if nums[j] < nums[i]:
                    dp[i] = max(dp[i], dp[j] + 1)
        return max(dp) if dp else 0
```

```python
def add_cont_change_and_max_returns(df):
    cont_change = list()
    max_returns = list()
    pct_returns = list()
    for index, row in df.iterrows():
        temp_val = list()
        temp_val.append(row['1_lag'])
        temp_val.append(row['2_lag'])
        temp_val.append(row['3_lag'])
        temp_val.append(row['4_lag'])
        temp_val.append(row['5_lag'])
        temp_val.append(row['6_lag'])
        temp_val.append(row['7_lag'])
        temp_val.append(row['8_lag'])
        temp_val.append(row['9_lag'])
        temp_val.append(row['10_lag'])
        temp_val.append(row['11_lag'])
        temp_val.append(row['12_lag'])
        temp_val.append(row['13_lag'])
        temp_val.append(row['14_lag'])
        temp_val.append(row['15_lag'])
        temp_val.append(row['16_lag'])
        temp_val.append(row['17_lag'])
        temp_val.append(row['18_lag'])
        temp_val.append(row['19_lag'])
        temp_val.append(row['20_lag'])


        cont_change.append(lengthOfLIS(temp_val))

        max_change = -1e9
        for i in range(len(temp_val)):
            for j in range(i, len(temp_val)):
                diff = temp_val[j] - temp_val[i]
                max_change = max(max_change, diff)
        max_returns.append(max_change)
        pct_returns.append(max_change * 100 / row['20_lag'])
    df['cont_change'] = cont_change
    df['max_returns'] = max_returns
    df['pct_returns'] = pct_returns
    return df
```

```python
for stocks in os.listdir(all_stocks_path):
    stock = all_stocks_path + stocks
    df = pd.read_csv(stock)
    df = add_prev_close_prices(df)
    df = add_cont_change_and_max_returns(df)
    df.to_csv(stock)
    print(stock)
```
