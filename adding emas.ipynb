```python
import pandas as pd
from datetime import date
from nsepy import get_history
import os

from pathlib import Path
proj_dir = str(Path.cwd())
```

```python
proj_dir
```

```python
all_stocks_path = proj_dir + "\\data\\historical\\"
```

```python
def get_ema(df, days):
    return df.ewm(span=days).mean()
```

```python
for stock in os.listdir(all_stocks_path):
    stock_file = all_stocks_path + stock
    print(stock_file)
    df = pd.read_csv(stock_file)

    if df.isnull().sum().sum() > 0:
        df.dropna()
    try:
        df.drop(["Symbol", "Series", "Turnover", "Trades", "Deliverable Volume", "%Deliverble"], axis = 1, inplace = True)
    except:
        pass
    ema_20 = list(get_ema(df['Close'], 20).values)
    ema_50 = list(get_ema(df['Close'], 50).values)
    ema_100 = list(get_ema(df['Close'], 100).values)
    ema_200 = list(get_ema(df['Close'], 200).values)
    
    df['ema20'] = ema_20
    df['ema50'] = ema_50
    df['ema100'] = ema_100
    df['ema200'] = ema_200
    print("EMAs added in " + stock_file)
    df.to_csv(stock_file)
```
